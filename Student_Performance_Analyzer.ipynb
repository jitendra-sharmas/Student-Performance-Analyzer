```python
df['Total_Marks'] = df['Maths'] + df['Physics'] + df['Chemistry']

df.head()
```

```python
df['Average_Marks'] = df['Total_Marks'] / 3

df.head()
```

```python
top10 = df.sort_values(by='Total_Marks', ascending=False)

top10.head(10)
```

```python
df['Chemistry'].max()
```

```python
def grade(avg):

    if avg >= 90:
        return 'A'

    elif avg >= 75:
        return 'B'

    elif avg >= 50:
        return 'C'

    else:
        return 'D'
```

```python
df['Grade'] = df['Average_Marks'].apply(grade)

df.head()
```

```python
df['Grade'].value_counts()
```

```python
df.groupby('Grade')['Average_Marks'].mean()
```
